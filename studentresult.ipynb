```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Expanded_data_with_more_features.csv')
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.pairplot(df)
```

```python
gb=df.groupby('ParentEduc').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("RELATION WITH PARENT EDUCATION")
plt.show()
```

```python
gb1=df.groupby('TestPrep').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("RELATION WITH TEST PREPARATION COURSE")
plt.show()
```

```python
gb2=df.groupby('ParentMaritalStatus').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
sns.heatmap(gb2,annot=True)
plt.title("RELATION WITH PARENT MARITAL STATUS")
plt.show()
```

```python
groupA = df.loc[(df["EthnicGroup"]=="group A")].count()
groupB = df.loc[(df["EthnicGroup"]=="group B")].count()
groupC = df.loc[(df["EthnicGroup"]=="group C")].count()
groupD = df.loc[(df["EthnicGroup"]=="group D")].count()
groupE = df.loc[(df["EthnicGroup"]=="group E")].count()
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]  # Counts of each group
l = ["Group A", "Group B", "Group C", "Group D", "Group E"]
print(mlist)
plt.pie(mlist, labels=l, autopct='%1.2f%%')
plt.title('Distribution of Ethnic Group')
plt.show()
```
