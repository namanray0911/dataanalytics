EXPLORING DATA ANALYSIS

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Size')
ax.bar_label(ax.containers[0])
plt.show()
```

THE BAR CHART SHOWS US THE MOST ORDERED SIZE IS M

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
S_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size',y='Qty', data=S_Qty)
```

THIS BARPLOT SHOWS THE QUANTITY OF DIFFRENT SIZES

```python
sns.countplot(data=df, x='Courier Status')
```

THIS COUNT PLOT SHOWS US ABOUT THE STATUS OF ORDER

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

THIS CHART SHOWS US VARIOUS TYPES OR ORDER



```python
# Checking B2B Data  by using pie chart
B2B_Check = df['B2B'].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

THIS PIE CHART SHOWS US THE PERCENTAGE OF RETAIL AND B2B BUYERS

```python

x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data, y_data)
plt.xlabel('Category ')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()

```

THIS SCATTER PLOT SHOWS RELATION BETWEEN CATEGORY OF ORDER AND SIZE

```python
plt.figure(figsize=(15, 9))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

THIS BAR CHART SHOWS NUMBER OF ORDERS FROM EACH STATE AND MOST BUYERS ARE FROM MAHARASTRA

CONCLUSION- HENCE THE DATA SUGGESTS MOST ORDER COMES FOR T SHIRTS .MOST ORDERED SIZE IS M AND MAHARASTRA IS BIGGEST BUYER OF ORDERS.
